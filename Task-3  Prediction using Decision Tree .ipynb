 # One-Hot Encoding
 
 **ML alogorithm do not work on categorical data so first we need to converts our categorical data in numercial data. so we use One hot encoding technique.**

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
oh_enc=OneHotEncoder(sparse=False)
```

```python
oh_enc_arr=oh_enc.fit_transform(df[["Species"]])
oh_enc_arr
```

```python
on_enc_df=pd.DataFrame(oh_enc_arr,columns=["Iris-setosa","Iris-versicolor","Iris-virginica"])   # Converting array in Dataframe
final_df=pd.merge(df,on_enc_df,left_index=True,right_index=True)                                #merging both dataframe
final_df.drop(["Species",],axis=1,inplace=True)                                             #removing unnecessary columns
```

```python
final_df
```

**Split Data**

```python
X=final_df.iloc[:,0:4]
Y=final_df.iloc[: ,4:7]
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.20,random_state=0)

print("Shape of X_train",X_train.shape)
print("Shape of X_test",X_test.shape)
print("Shape of Y_train",Y_train.shape)
print("Shape of Y_test",Y_test.shape)
```

**Decision Tree Classification**

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
classifier=DecisionTreeClassifier(criterion="gini")
classifier.fit(X_train,Y_train)
```

```python
y_predict=classifier.predict(X_test)
```

# Prediction

```python
SepalLength=float(input("Enter SepalLengthCm  (0.1-8.0)-->"))
SepalWidth=float(input("Enter SepalWidthCm (0.1-4.5) -->"))
PetalLength=float(input("Enter PetalLengthCm (0.1-7.0)-->"))
SepalWidth=float(input("Enter PetalWidthCm  (0.1-3.0)-->"))
```

```python
result=classifier.predict([[SepalLength,SepalWidth,PetalLength,SepalWidth]])
```

```python
if result[0][0]==1:
    print("Iris-Setosa")
elif result[0][1]==1:
    print("Iris-versicolor")
else:
        print("Iris-Virginica")
```

```python
classifier.score(X_test,Y_test)
```
